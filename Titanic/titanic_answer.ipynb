```python
from tensorflow.keras import models, layers
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python
raw_titanic = pd.read_csv('data/train.csv')
```

```python
raw_titanic.drop(['Name', 'Ticket', 'Fare', 'Cabin', 'Embarked'], axis=1, inplace=True)
raw_titanic.head()
```

```python
titanic_data = pd.get_dummies(raw_titanic, columns=['Sex'], dtype='int')
titanic_data.head()
```

```python
titanic_data['age_label'] = pd.cut(titanic_data.Age,
                                   bins=[0, 2, 17, 65, 99],
                                   labels=['Baby', 'Child', 'Adult', 'Elderly'])
titanic_data.head()
```

```python
titanic_data.fillna('Adult', inplace=True)
titanic_data.head(20)
```

```python
titanic_data = pd.get_dummies(titanic_data, columns=['age_label'], dtype='int')
titanic_data.head()
```
