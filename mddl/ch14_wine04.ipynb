```python
from tensorflow.keras import models, layers
from sklearn.model_selection import train_test_split
import pandas as pd
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
```

```python
df = pd.read_csv('data/wine.csv', header=None)
X = df.iloc[:, :12]
y = df.iloc[:, 12]
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)

model = models.Sequential()
model.add(layers.Dense(30, input_dim=12, activation='relu'))
model.add(layers.Dense(12, activation='relu'))
model.add(layers.Dense(8, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=20)
modelpath = 'model/wine-best.hdf5'
checkpoint = ModelCheckpoint(filepath=modelpath, verbose=0, monitor='val_loss', save_best_only=True)
```

```python
history = model.fit(X_train, y_train, epochs=2000, batch_size=500, verbose=1, validation_split=0.25, callbacks=[checkpoint, early_stopping])
```

```python
score = model.evaluate(X_test, y_test)
print('Test Accuracy: ', score[1])
```
